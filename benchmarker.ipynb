```python
import matplotlib.pyplot as plt

class benchmark:
    
    def __init__(self, model_name, framework, video_name, latency, object_detection):
        self.model_name = model_name
        self.framework = framework
        self.video_name = video_name
        self.latency = latency
        self.object_detection = object_detection

    def __init__(self, filename, name):
        # split = filename.replace(".json", "").split("_")
        # split_len = len(split)
        # self.framework = split[0]
        self.video_name = filename.replace(".json", "")
        self.name = name
        # for i in range(1, split_len-1):
        #     self.model_name.append(split[i])

    def set_latency(self, latency):
        self.latency = latency
        # if (self.framework == "argus"):
        #     self.latency = [i/1000 for i in self.latency]

    def set_object_detection(self, object_detection):
        self.object_detection = object_detection

    def get_model_name(self):
        name = ""
        if len(self.model_name) > 1:
            for i in self.model_name:
                name += i + "+"
        else:
            name = self.model_name[0]

        return name

    def plot_graph(self):
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))
        fig.suptitle(self.framework + " " + self.get_model_name() + " " + self.video_name + ' Latency vs Frame #')
        ax1.plot(self.latency)
        ax2.plot(self.latency[1:])
        ax1.set(xlabel='Frame #', ylabel='Latency')
        ax1.grid()
        ax2.grid()
        # fig.show()
```

```python
import os
import json

BENCHMARK_FOLDER = "./benchmark_results"

def discover_benchmarks():
    benchmarks = []

    for files in os.listdir(BENCHMARK_FOLDER):
        file_path = os.path.join(BENCHMARK_FOLDER, files)
        if "json" not in files:
            continue

        data = open(file_path, "r")
        json_data = json.load(data)

        benchmark_data = benchmark(files)
        benchmark_data.set_latency(json_data['latency'])
        benchmark_data.set_object_detection(json_data['object_detection'])

        benchmarks.append(benchmark_data)
        video_to_benchmarks[benchmark_data.video_name].append(benchmark_data)

    return benchmarks


benchmarks = discover_benchmarks()
```
